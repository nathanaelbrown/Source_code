```python
#local library to read pdfs
import pdf_to_text as pt

#import to work with file directory
import glob
import os

#searches strings
import re

#load path
path = '../RA_1/'
```

```python
#Returns the entire list of files in folder RA_1
list_of_files = glob.glob(path + '*') # * means all if need specific format then *.pdf

#sort list delete nonsearchable files
print("Initial number of files", len(list_of_files))
#print(type(list_of_files))

#This code deletes the duplicate unsearchable.pdf
for i in list_of_files:
    if '_searchable' in i:
        delete_file = i.replace('_searchable.pdf', '')
        #print(delete_file)
        try:
            list_of_files.remove(delete_file)
        except:
            print(i, ' is unique')

print('Number of files after deleting unsearchable duplicates', len(list_of_files))
# for j in list_of_files:
#     print(j)
```

```python
#Empty List for testing results
l=[]
count=0

for k in list_of_files:
    x = ''
    try:
        x=pt.pdf_to_text(k)
    except:
        print(k, 'failed to load')
        
    strx = str(x)
    if 'Entity #: ' in strx:
        #s = 'asdf=5;iwantthis123jasd'
        count+=1
        print(strx)
        result = re.search("Entity #:(.*?)n", strx)
        id=result.group(1).strip(" \\")
        print()
        print('Entry: ', count)
        print('Entity is:', id)
        if 'Name: ' in strx:
            result = re.search("Name: (.*?)\\\\", strx)
            len_groups = len(result.groups())
            #print("Number of results: ", len_groups)
            incoporator = result.group(len_groups).strip(" \\")
            print('Incorporator is: ', incoporator)
        else: print('Not Found')
        print()
        #y=input('Continue?')
        #if y=='No': break
        l.append(k)

print('Number of files that were processed', len(l)) 
print(l)      
```
