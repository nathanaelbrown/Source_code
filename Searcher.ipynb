```python
###########Search and Hightlight######################
#
#  When prompted, enter a search term to return a list of files that contain the term with 
#  the term hightlighted in the text.
#
#####################################################


#local library to read pdfs
import pdf_to_text as pt

#import to work with file directory
import glob
import os

#searches strings
import re
import pandas as pd

#load path
path = '../RA_1/'

#Hightlighter function
def highlight_one(keyword, text):
    #print(keyword)
    #print(type(keyword))
    replacement = "\u001b[43m" + keyword + "\u001b[0m"#"\033[1m\033[91m" + keyword + "\033[39m"
    text = re.sub(keyword, replacement, text, flags=re.I)
    return text
```

```python
#Returns the entire list of files in folder RA_1
list_of_files = glob.glob(path + '*') # * means all if need specific format then *.pdf

#sort list delete nonsearchable files
print("Initial number of files", len(list_of_files))
#print(type(list_of_files))

#This code deletes the duplicate unsearchable.pdf
for i in list_of_files:
    if '_searchable' in i:
        delete_file = i.replace('_searchable.pdf', '')
        #print(delete_file)
        try:
            list_of_files.remove(delete_file)
        except:
            print(i, ' is unique')

print('Number of files after deleting unsearchable duplicates', len(list_of_files))
# for j in list_of_files:
#     print(j)
```
