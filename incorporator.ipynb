```python

#local library to read pdfs
import pdf_to_text as pt

#import to work with file directory
import glob
import os

#searches strings
import re
import pandas as pd

#load path
path = '../RA_1/'

#Hightlighter function
def highlight_one(keyword, text):
    replacement = "\u001b[43m" + keyword + "\u001b[0m"
    text = re.sub(keyword, replacement, text, flags=re.I)
    return text
```

```python
#Returns the entire list of files in folder RA_1
list_of_files = glob.glob(path + '*.pdf') # * means all if need specific format then *.pdf

#sort list delete nonsearchable files
print("Initial number of files", len(list_of_files))
#print(type(list_of_files))

#This code deletes the duplicate unsearchable.pdf
for i in list_of_files:
    if '_searchable' in i:
        delete_file = i.replace('_searchable.pdf', '')
        #print(delete_file)
        try:
            list_of_files.remove(delete_file)
        except:
            print(i, ' is unique')

print('Number of files after deleting unsearchable duplicates', len(list_of_files))
# for j in list_of_files:
#     print(j)
```

```python
#Empty List for testing results
l=[]
count=0
df=pd.DataFrame([],columns=['Company ID', 'Coder', 'Incorporator 1'])
#list_of_files = ['../RA_1/10121437_1-08-2020_Creation Filing.pdf']

for k in list_of_files:
    x = ''
    try:
        x=pt.pdf_to_text(k)
    except:
        print(k, 'failed to load')
        
    strx = str(x)
    
    pattern = re.compile(r"Entity  ?#:(.*?)\\")
   
    if re.search(pattern, strx):
        count+=1
        result=re.search(pattern, strx)
        #result = re.search(r"Entity  ?#:(.*?)\\\\", strx)
       
        eid=result.group(1).strip(r" ")
        print()
        print('Entry: ', count)
        print('Entity ID# is:', eid)
         
        incorppattern1 = re.compile(r"Incorporator\(s\) \\n\\n(.*?)\\")
        incorppattern2 = re.compile(r"Name:(.*?)\\")
        
        if re.search(incorppattern1, strx):
            result = re.findall(incorppattern1, strx)[-1]
            incorporator = result.strip(r" ")
            print('Incorporator is: ', incorporator)
            df=df.append(pd.DataFrame([[eid, 'Nate Py_Script', incorporator]], columns = ['Company ID', 'Coder', 'Incorporator 1']), ignore_index=True)
            #print(pd.DataFrame([[eid, incorporator]], columns = ['Entity', 'incorporator']).head())
            #print(df.head())
            print()
            print(highlight_one(incorporator, strx)) 
        elif re.search(incorppattern2, strx):
            #result = re.findall("Name:(.*?)\\\\", strx)[-1]
            result = re.findall(incorppattern2, strx)[-1]
            incorporator = result.strip(r" ")
            print('Incorporator is: ', incorporator)
            df=df.append(pd.DataFrame([[eid, 'Nate Py_Script', incorporator]], columns = ['Company ID', 'Coder', 'Incorporator 1']), ignore_index=True)
            #print(pd.DataFrame([[eid, incorporator]], columns = ['Entity', 'incorporator']).head())
            #print(df.head())
            print()
            print(highlight_one(incorporator, strx))
            #xyz=input('Continue?')
            #if xyz=='No': break
        else: print('Incorporator Not Found')
            
        if k.count(str(eid)) == 0:
            print('WARNING: Entity #', eid, 'does not match file name', k)
        print()
          
        l.append(k)
    else: 
        print()
        print('Entity Not Found for File: ', k, 'Not searched')
        print()
        print(strx)
    
    #y=input('Continue?')
    #if y=='No': break   
print('Number of files that were processed', len(l)) 
print(df.head())
print(df.tail())
```
